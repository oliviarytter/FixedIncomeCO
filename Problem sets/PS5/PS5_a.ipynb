## Problem 1 - Black-Scholes Formula Derivation

**Setup:** Suppose $S_t$ follows a geometric Brownian motion with dynamics
$$
dS_t = \mu S_t dt + \sigma S_t dW_t, \quad S_0 = s_0
$$
where $\mu$ and $\sigma$ are constants. Also assume there exists a risk-free bank account $B_t$ with constant interest rate $r$ and dynamics
$$
dB_t = rB_t dt, \quad B_0 = b_0
$$

---



### 1.a) Show that under the risk neutral measure, the dynamics of $S_t$ must be $dS_t = rS_t dt + \sigma S_t dW_t$

Under the risk neutral measure $\mathbb{Q}$, the ratio $Z_t = \frac{S_t}{B_t}$ must be a martingale. Let us find the $\mathbb{Q}$-dynamics of this ratio using Ito's formula.

**Applying Ito to $Z_t = \frac{S_t}{B_t}$:**

We need:
$$
\frac{\partial Z}{\partial S} = \frac{1}{B_t}, \quad \frac{\partial Z}{\partial B} = -\frac{S_t}{B_t^2}, \quad \frac{\partial^2 Z}{\partial B^2} = \frac{2S_t}{B_t^3}
$$

Under the risk-neutral measure, assume $dS_t = rS_t dt + \sigma S_t dW_t$. Then:
$$
dZ_t = \frac{1}{B_t}dS_t - \frac{S_t}{B_t^2}dB_t + \frac{S_t}{B_t^3}(dB_t)^2
$$

Substituting:
$$
dZ_t = \frac{1}{B_t}(rS_t dt + \sigma S_t dW_t) - \frac{S_t}{B_t^2}(rB_t dt) + \frac{S_t}{B_t^3}r^2B_t^2(dt)^2
$$

Since $(dt)^2 = 0$:
$$
dZ_t = \frac{rS_t}{B_t}dt + \frac{\sigma S_t}{B_t}dW_t - \frac{rS_t}{B_t}dt = \frac{\sigma S_t}{B_t}dW_t
$$

This is clearly a martingale because it has zero drift and is purely driven by the Brownian motion $dW_t$.

**Verifying the martingale property:**

Integrating from $t$ to $T$:
$$
Z_T = Z_t + \int_t^T \frac{\sigma S_u}{B_u}dW_u
$$

Taking conditional expectations:
$$
\mathbb{E}[Z_T \mid \mathcal{F}_t] = Z_t + \mathbb{E}\left[\int_t^T \frac{\sigma S_u}{B_u}dW_u \mid \mathcal{F}_t\right] = Z_t
$$

Therefore, $Z_t = \frac{S_t}{B_t}$ is indeed a martingale under $\mathbb{Q}$ when the stock has drift $r$.

*Intuition: Under the risk-neutral measure, all assets when normalized by the risk-free asset become martingales. This means they have no "excess drift" beyond the risk-free rate. The key insight is that the physical drift $\mu$ is replaced by the risk-free rate $r$ under $\mathbb{Q}$, while the volatility $\sigma$ remains unchanged. This is the fundamental principle of risk-neutral pricing.*

### 1.b) Show that the solution $S(T)$ corresponding to the $\mathbb{Q}$-dynamics is $S(T) = s_0 \exp\left[\left(r - \frac{1}{2}\sigma^2\right)T + \sigma W(T)\right]$

Apply Ito's formula to $X_t = \ln(S_t)$ where $dS_t = rS_t dt + \sigma S_t dW_t$.

**Computing partial derivatives:**
$$
\frac{\partial f}{\partial S} = \frac{1}{S}, \quad \frac{\partial^2 f}{\partial S^2} = -\frac{1}{S^2}
$$

**Applying Ito's formula:**
$$
dX_t = \frac{1}{S_t}dS_t + \frac{1}{2}\left(-\frac{1}{S_t^2}\right)(dS_t)^2
$$

Substituting:
$$
dX_t = \frac{1}{S_t}(rS_t dt + \sigma S_t dW_t) - \frac{1}{2S_t^2}(\sigma S_t)^2 dt
$$
$$
= r dt + \sigma dW_t - \frac{1}{2}\sigma^2 dt = \left(r - \frac{1}{2}\sigma^2\right)dt + \sigma dW_t
$$

**Integrating from $0$ to $T$:**
$$
X_T = X_0 + \int_0^T \left(r - \frac{1}{2}\sigma^2\right)ds + \int_0^T \sigma dW_s
$$
$$
X_T = x_0 + \left(r - \frac{1}{2}\sigma^2\right)T + \sigma W_T
$$

Since $X_t = \ln(S_t)$, we have $X_0 = \ln(s_0)$:
$$
\ln(S_T) = \ln(s_0) + \left(r - \frac{1}{2}\sigma^2\right)T + \sigma W_T
$$

Taking exponentials:
$$
S_T = s_0 \exp\left[\left(r - \frac{1}{2}\sigma^2\right)T + \sigma W_T\right]
$$

*Intuition: This is the same GBM solution as before, but with $\mu$ replaced by $r$. The $-\frac{1}{2}\sigma^2$ correction term still appears due to Ito's lemma, ensuring that $\mathbb{E}^{\mathbb{Q}}[S_T] = s_0 e^{rT}$ under the risk-neutral measure.*

### 1.c) Show that $X(T) \sim \mathcal{N}\left(x_0 + \left(r - \frac{1}{2}\sigma^2\right)T, \sigma^2 T\right)$ and find the PDF

From part (b), we have:
$$
X_T = x_0 + \left(r - \frac{1}{2}\sigma^2\right)T + \sigma W_T
$$

Since $W_T \sim \mathcal{N}(0, T)$ and $X_T$ is a constant plus a normal random variable:

**Mean:**
$$
\mathbb{E}[X_T] = x_0 + \left(r - \frac{1}{2}\sigma^2\right)T + \sigma \mathbb{E}[W_T] = x_0 + \left(r - \frac{1}{2}\sigma^2\right)T
$$

**Variance:**
$$
\text{Var}[X_T] = \sigma^2 \text{Var}[W_T] = \sigma^2 T
$$

Therefore:
$$
X_T \sim \mathcal{N}\left(x_0 + \left(r - \frac{1}{2}\sigma^2\right)T, \sigma^2 T\right)
$$

**Probability density function:**

The PDF of a normal random variable $Y \sim \mathcal{N}(\mu, \sigma^2)$ is:
$$
f_Y(y) = \frac{1}{\sqrt{2\pi\sigma^2}}\exp\left(-\frac{(y-\mu)^2}{2\sigma^2}\right)
$$

For $X_T$:
$$
f_{X(T)}(x) = \frac{1}{\sqrt{2\pi\sigma^2 T}}\exp\left(-\frac{\left(x - x_0 - \left(r - \frac{1}{2}\sigma^2\right)T\right)^2}{2\sigma^2 T}\right)
$$

*Intuition: Since $S_T$ is log-normally distributed, its logarithm $X_T = \ln(S_T)$ is normally distributed. This is fundamental to option pricing because we can now compute expectations involving $S_T$ using the normal distribution of $X_T$.*

### 1.d) Argue that the time 0 price of the call option can be found from the given expression

A European call option has payoff at maturity:
$$
C(S_T) = \max(S_T - K, 0) = (S_T - K)^+
$$

**Risk-neutral pricing formula:**

By the fundamental theorem of asset pricing, the time-0 price is:
$$
C = e^{-rT}\mathbb{E}^{\mathbb{Q}}[\max(S_T - K, 0)]
$$

**Expressing in terms of $X_T = \ln(S_T)$:**

Since $S_T = e^{X_T}$, we can write:
$$
C = e^{-rT}\mathbb{E}^{\mathbb{Q}}[(e^{X_T} - K)^+]
$$

For a continuous random variable with PDF $f_{X(T)}(x)$:
$$
C = e^{-rT}\int_{-\infty}^{\infty} (e^x - K)^+ f_{X(T)}(x)dx
$$

**Finding the integration limits:**

The payoff $(e^x - K)^+ > 0$ only when $e^x > K$, i.e., $x > \ln K$. Therefore:
$$
C = e^{-rT}\int_{\ln K}^{\infty} (e^x - K) f_{X(T)}(x)dx
$$

**Splitting into two integrals:**
$$
C = e^{-rT}\int_{\ln K}^{\infty} e^x f_{X(T)}(x)dx - Ke^{-rT}\int_{\ln K}^{\infty} f_{X(T)}(x)dx = I_1 + I_2
$$

*Intuition: We've transformed the option pricing problem into evaluating two integrals. $I_1$ represents the present value of receiving the stock when it expires in-the-money, while $I_2$ represents the present value of paying the strike price. The integration starts at $\ln K$ because that's where the option becomes valuable.*

---

### 1.e) Show that this results in the Black-Scholes formula

Let $k = \ln K$ and $\mu_0 = x_0 + \left(r - \frac{1}{2}\sigma^2\right)T$.

**Computing $I_1$:**
$$
I_1 = e^{-rT}\int_k^{\infty} e^x \frac{1}{\sqrt{2\pi\sigma^2 T}}\exp\left(-\frac{(x - \mu_0)^2}{2\sigma^2 T}\right)dx
$$

Combining exponents:
$$
= \frac{e^{-rT}}{\sqrt{2\pi\sigma^2 T}}\int_k^{\infty} \exp\left(-\frac{x^2 - 2x\mu_0 + \mu_0^2 - 2x\sigma^2 T}{2\sigma^2 T}\right)dx
$$

**Completing the square:**

The exponent can be written as:
$$
-\frac{x^2 - 2x(\mu_0 + \sigma^2 T) + \mu_0^2}{2\sigma^2 T} = -\frac{[x - (\mu_0 + \sigma^2 T)]^2 - 2\mu_0\sigma^2 T - \sigma^4 T^2}{2\sigma^2 T}
$$

Therefore:
$$
I_1 = e^{-rT + \mu_0 + \frac{1}{2}\sigma^2 T} \frac{1}{\sqrt{2\pi\sigma^2 T}}\int_k^{\infty} \exp\left(-\frac{[x - (\mu_0 + \sigma^2 T)]^2}{2\sigma^2 T}\right)dx
$$

**Substitution:**

Let $z = \frac{x - \mu_0 - \sigma^2 T}{\sigma\sqrt{T}} = \frac{x - x_0 - (r + \frac{1}{2}\sigma^2)T}{\sigma\sqrt{T}}$

Then $dx = \sigma\sqrt{T}dz$ and when $x = k$:
$$
z^* = \frac{k - x_0 - (r + \frac{1}{2}\sigma^2)T}{\sigma\sqrt{T}} = -d_1
$$

where:
$$
d_1 = \frac{\ln\left(\frac{s_0}{K}\right) + \left(r + \frac{1}{2}\sigma^2\right)T}{\sigma\sqrt{T}}
$$

Therefore:
$$
I_1 = e^{x_0}\int_{-\infty}^{d_1} \frac{1}{\sqrt{2\pi}}e^{-\frac{z^2}{2}}dz = s_0\Phi(d_1)
$$

**Computing $I_2$:**
$$
I_2 = -Ke^{-rT}\int_k^{\infty} \frac{1}{\sqrt{2\pi\sigma^2 T}}\exp\left(-\frac{(x - \mu_0)^2}{2\sigma^2 T}\right)dx
$$

**Substitution:**

Let $z = \frac{x - \mu_0}{\sigma\sqrt{T}} = \frac{x - x_0 - (r - \frac{1}{2}\sigma^2)T}{\sigma\sqrt{T}}$

When $x = k$:
$$
z^* = \frac{k - x_0 - (r - \frac{1}{2}\sigma^2)T}{\sigma\sqrt{T}} = -d_2
$$

where:
$$
d_2 = \frac{\ln\left(\frac{s_0}{K}\right) + \left(r - \frac{1}{2}\sigma^2\right)T}{\sigma\sqrt{T}} = d_1 - \sigma\sqrt{T}
$$

Therefore:
$$
I_2 = -Ke^{-rT}\int_{-\infty}^{d_2} \frac{1}{\sqrt{2\pi}}e^{-\frac{z^2}{2}}dz = -Ke^{-rT}\Phi(d_2)
$$

**Black-Scholes formula:**
$$
C = I_1 + I_2 = s_0\Phi(d_1) - Ke^{-rT}\Phi(d_2)
$$

where:
$$
d_1 = \frac{\ln\left(\frac{s_0}{K}\right) + \left(r + \frac{1}{2}\sigma^2\right)T}{\sigma\sqrt{T}}, \quad d_2 = d_1 - \sigma\sqrt{T}
$$

*Intuition: The Black-Scholes formula has an elegant interpretation: $s_0\Phi(d_1)$ is the present value of receiving the stock (weighted by the risk-neutral probability of finishing in-the-money), and $Ke^{-rT}\Phi(d_2)$ is the present value of paying the strike (also weighted by a related probability). The difference $d_1 - d_2 = \sigma\sqrt{T}$ represents the volatility effect over the option's life. This formula revolutionized finance by providing a closed-form solution for option prices.*
