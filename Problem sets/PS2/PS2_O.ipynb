## Problem 1 - Interest rate payer swap
Vi betragter en interest rate payer swap med:
- **Fixed leg**: Betaler kuponer af størrelse R, I gange om året
- **Floating leg**: Betaler simple LIBOR rate kuponer, J gange om året
- **Løbetid**: N år
- **Principal**: K
- **M = J/I** (antal floating betalinger per fixed betaling)
- **Floating kupondatoer**: $T₁, T₂, ..., T_{JN}$
- **Fixings**: Annonceret på tidspunkter $T₀, T₁, ..., T_{JN-1}$
- **LIBOR rate**: $L(Tᵢ, Tᵢ₊₁)$ annonceret på tidspunkt Tᵢ, betalt på tidspunkt $Tᵢ₊₁$



### Problem 1.a - Find an expression for the value of the fixed leg.
The fixed leg pays I times a year, matures in N years, meaning that it pays IN in total. The payment dates are: $T_M, T_{2M}, T_{3M}, ..., T_{IN·M}$.

PV of the fixed leg is:

$$PV_{fixed} = \sum_{i=1}^{IN} p(t, T_{iM}) \cdot \alpha_{iM} \cdot R \cdot K$$

Where $α_{iM}=T_{iM} - T_{(i-1)M}$ is the time between coupons.
- p(t, T_{iM}) are the zero coupon prices
- R is the coupon payment
- K is the principal 

If we define 
$$S_0 = \sum_{i=1}^{IN} p(t, T_{iM}) \cdot (T_{iM} - T_{(i-1)M})$$

We get
$$\boxed{PV_{fixed} = S_0 \cdot R \cdot K}$$

**Explanation**: Fixed leg is the sum of discounted fixed leg payments

---

### Problem 1.b - Find an expression for the value of the floating leg
Floating leg pays LIBOR coupons J times a year in N years, equal to JN payments in total. For a single floating coupon paid at $T_i$ based on a fixing in $T_{i-1}$

$$PV_i = p(t, T_i) \cdot (T_i - T_{i-1}) \cdot L(T_{i-1}, T_i) \cdot K$$
- p(t,T_i) is the ZCB price at the given time 
- L(T_{i-1}, T_i) is the LIBOR rate

The LIBOR rate is defined as:
$$L(T_{i-1}, T_i) = \frac{1}{T_i - T_{i-1}} \left(\frac{1}{p(T_{i-1}, T_i)} - 1\right)$$

At the value of time $T=T_{i-1}$ (K is temporarily 1)
$$p(T_{i-1}, T_i) \cdot (T_i - T_{i-1}) \cdot L(T_{i-1}, T_i)$$
Inserting LIBOR:
$$= p(T_{i-1}, T_i) \cdot (T_i - T_{i-1}) \cdot \frac{1}{T_i - T_{i-1}} \left(\frac{1}{p(T_{i-1}, T_i)} - 1\right)$$
$$ = 1 - p(T_{i-1}, T_i)$$

Discounting back to $T_{i-1}=t$, we first multiply with the ZCB price for time, t:
$$PV_i = p(t,T_{i-1})\cdot K [1 - p(T_{i-1}, T_i)]$$
Multiplying into the paranthesis:
$$PV_i = K \cdot [p(t,T_{i-1}) - p(t,T_{i-1}) \cdot p(T_{i-1}, T_i)]$$
Using the rule: $p(t,T_{i-1}) \cdot p(T_{i-1}, T_i)=p(t,T_i)$

$$PV_i = K \cdot [p(t, T_{i-1}) - p(t, T_i)]$$

The total value is the sum of the floating legs throughout the period:
$$PV_{float} = \sum_{i=1}^{JN} K \cdot [p(t, T_{i-1}) - p(t, T_i)]$$

If we wrote out the sum terms would cancel and we would end up with (teleskopic sum):
$$\boxed{PV_{float} = K \cdot [p(t, T_0) - p(t, T_{JN})]}$$

**Intuition**: 
- $p(t, T₀)$ = ZCB price of fixing at time t = 0
- $p(t, T_{JN})$ = ZCB price for last cashflow of floating leg

---

### Problem 1.c - Find an expression for the par swap rate 
**Identify for each of the elements of the expression, if they originate from the fixed or the floating leg.**

The par swap rate is the fixed rate, $R^*$ that makes the swap equal to zero at issuance (t=0).

Setting $PV_{float} = PV_{fixed}$ at time $t=0$:
$$K \cdot [p(0, T_0) - p(0, T_{JN})] = S_0 \cdot R^* \cdot K$$

Solving for $R^*$:
$$R^* = \frac{K \cdot [p(0, T_0) - p(0, T_{JN})]}{S_0 K}$$
Inserting $S_0=\sum_{i=1}^{IN} p(t, T_{iM}) \cdot (T_{iM} - T_{(i-1)M})$:
$$R^* = \frac{p(0, T_0) - p(0, T_{JN})}{S_0}$$

$$\boxed{R^* = \frac{p(0, T_0) - p(0, T_{JN})}{\sum_{i=1}^{IN} p(0, T_{iM}) \cdot (T_{iM} - T_{(i-1)M})}}$$

**Terms**:
- **Denominator** (from fixed leg):
  - Sum of the discounted accrual factors for the fixed leg payments. $S_0$ is the accrual factor! Comes solely from the fixed leg.

---

### Problem 1.d - Find accrual factor
The accrual factor comes from the denominator of the par swap rate found above

Let us define the accrual factor for a given period using the formula for the par swap rate:
$$S_0 = \frac{1}{\sum_{i=1}^{IN} p(0, T_{iM}) \cdot (T_{iM} - T_{(i-1)M})}$$

---

### Problem 1.e - Compare the par swap rate and the accrual factor you found to the case where coupons to both the fixed and floating legs occur at the same time.

**Case 1: Different frequency** (our case)
$$R^* = \frac{p(0, T_0) - p(0, T_{JN})}{\sum_{i=1}^{IN} p(0, T_{iM}) \cdot (T_{iM} - T_{(i-1)M})}$$

**Case 2: Same frequency** (I = J, M = 1)
$$R^*_{sync} = \frac{p(0, T_0) - p(0, T_N)}{\sum_{i=1}^{N} p(0, T_i) \cdot (T_i - T_{i-1})}$$

**Numerator:**  
The numerator represents the value of the floating leg and is identical in both cases, as it telescopes to $[ p(0,T_0) - p(0,T_{\text{end}}) ]$ independently of payment frequency.

**Denominator (accrual factor):**  
Although the fixed leg has fewer payments when coupons occur less frequently, each accrual period is correspondingly longer. These effects offset each other, so both sums approximate the same discounted time integral.

**Conclusion:**  
The accrual factor and the par swap rate are therefore essentially unchanged whether fixed and floating coupons occur at the same time or at different frequencies.

## Swap after some time
Assume now that $T_J < t < T_{J+1}$, i.e. a little more than one year has past.
The fixing at time $T_J$ is already announced and the coupon $L(T_J,T_{J+1})$ to the floating leg at time $T_{J+1}$ is already known.

### Problem 1.f: Find an expression for the value of the fixed leg.

The value of the fixed leg is the sum of all future fixed payments at time t. All fixed payments before are already paid - value is 0.

- Floating pays at: $T_1, T_2,..,T_{JN}$
- Fixed pays every M floating date: $T_M, T_{2M}, T_{3M}...,$.

This means that the first fixed payment after $t$ is on the floating index: $J+M$. If J floating periods have already passed by, and we have fixed payments every M'th period. We must have that number of already paid fixed payments are: $J/M$. This is NOT included as it is already paid, meaning that the first future payment is $J/M+1$

$$\boxed{PV_{fixed}(t) = \sum_{i= J/M + 1}^{IN} p(t, T_{iM}) \cdot (T_{iM} - T_{(i-1)M}) \cdot R \cdot K}$$


---

### 1.g Find an expression for the value of the floating leg.
Floating leg has both known payments and unknown
1. **Known payments** $T_{J+1}: L(T_J, T_{J+1})$ is already fixed ($T_{J+1}$ is known from the inequality)
2. **Unknown future payments** From $T_{J+2}$ til $T_{JN}$

The cash flow at $T_{J+1}$ is known: $K(T_{J+1}-T_J)L(T_J,T_{J+1}). We multiply by the ZCB price at the time to get the value:
$$p(t, T_{J+1}) \cdot K(T_{J+1}-T_J) \cdot L(T_J,T_{J+1})$$
The values after $T_{J+1}$ are unknown, however we know that the value of a floating coupon in one period is:
$$PV_i = K \cdot [p(t, T_{i-1}) - p(t, T_i)]$$

So for all the remaining floating coupons we can write the value as:
$$\sum_{i=J+2}^{JN} K \cdot [p(t, T_{i-1}) - p(t, T_i)]$$

This sum is telescopic:
$$ K \cdot [p(t, T_{J+1}) - p(t, T_{J+2})] +K \cdot [p(t, T_{J+2}) - p(t, T_{J+3})]+..$$

Everything cancels except:
$$ K \cdot [p(t, T_{J+1}) - p(t, T_{JN})]$$

Now we combine the known and unknown parts:
$$PV_{float}(t) = p(t, T_{J+1}) \cdot K(T_{J+1}-T_J) \cdot L(T_J,T_{J+1})+ K \cdot [p(t, T_{J+1}) - p(t, T_{JN})]$$

Intuition: After time $t$, the floating leg consists of one known coupon plus a remaining par-resetting leg, which is why its value splits into a discounted known payment and a telescoping discount-factor difference.

---

### 1.h Find an expression for the par swap rate and the accrual factor.
The par swap rate at time $t$ is the fixed rate $R^*$ that makes the value of a new swap entered at time 
$t$ equal to zero. They set $PV_{float}=PV_{fixed}$

$$p(t, T_{J+1}) \cdot K(T_{J+1}-T_J) \cdot L(T_J,T_{J+1})+ K \cdot [p(t, T_{J+1}) - p(t, T_{JN})] = \sum_{i= J/M + 1}^{IN} p(t, T_{iM}) \cdot (T_{iM} - T_{(i-1)M}) \cdot R \cdot K$$

Isolate for $R^*$
$$R^* = \frac{p(t, T_{J+1}) \cdot (T_{J+1}-T_J) \cdot L(T_J,T_{J+1})+ \cdot [p(t, T_{J+1}) - p(t, T_{JN})]}{\sum_{i= J/M + 1}^{IN} p(t, T_{iM}) \cdot (T_{iM} - T_{(i-1)M})}$$

This is the par rate that re-values the existing swap at time $t$

---

### 1.i Find an expression for the PnL of the payer swap. Does the expression for the PnL depend on LIBOR?
$PnL = PV_{float}-PV_{fixed} $

As the accrual factor, $S_t=\sum_{i= J/M + 1}^{IN} p(t, T_{iM}) \cdot (T_{iM} - T_{(i-1)M})$
$$PV_{fixed}=S_t R K$$
We found the par swap rate in g) to be:
$$R^* = \frac{p(t, T_{J+1}) \cdot (T_{J+1}-T_J) \cdot L(T_J,T_{J+1})+ \cdot [p(t, T_{J+1}) - p(t, T_{JN})]}{\sum_{i= J/M + 1}^{IN} p(t, T_{iM}) \cdot (T_{iM} - T_{(i-1)M})}$$

Inserting $PV_{fixed}=S_t R K$ to find the par swap rate we get:
$$R^* = \frac{p(t, T_{J+1}) \cdot (T_{J+1}-T_J) \cdot L(T_J,T_{J+1})+ \cdot [p(t, T_{J+1}) - p(t, T_{JN})]}{S_t}$$

The par swap rate is the fixed rate $R^*$ that makes the value of a new swap entered at time $t$ equal to zero.

$$PV_{float}-PV_{fixed}=0$$
$$PV_{float}=S_t K R_t^*$$

$$ PnL=S_t K R_t^*-S_t K R_0^*=S_t K (R_t^*-R_0^*)$$

Intuition: The PnL of the payer swap depends on the principal K, the accrual factor $S_t$ and the change in the par swap rate.

## Problem 2 – Yield Curve Construction and Swap Valuation

Assume: 
- The year consists of 12 months each with exactly 30 days 
- ALl payments occur at the end of day on the last day of the month
- No credit risk and the the principal of all bonds is 100. 

**Today:** 2019-12-30. 
**LIBOR fixings:** 
- 3M LIBOR = 0.01570161, 
- 6M LIBOR = 0.01980204.

Bonds (coupon paid simple at stated frequency, principal 100):
| # | Type | Maturity | Coupon Freq | Annual Coupon | Price |
|---|------|----------|-------------|---------------|-------|
| i | 3Y Fixed Bullet | Dec 30, 2020 | Quarterly | 4% | 102.33689177 |
| ii | 5Y Fixed Bullet | Dec 30, 2020 | Semi-annual | 5% | 104.80430234 |

| vi | 30Y Fixed Bullet | Dec 30, 2021 | Annual | 3% | 101.82604116 |

We generate coupon schedules from 2019-12-30 to each maturity with end-of-month payments and 30/360 accruals. We include principal at maturity. The cashflow matrix has one column per unique payment date across all bonds; entries are the cashflows on that date (coupon + principal if maturity).

### 2.a) Cashflow matric 

Build the **cashflow matrix** for the given fixed-rate bullet bonds (using the exact coupon schedules).


```python
import numpy as np
import fixed_income_derivatives_E2025 as fid
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from datetime import date

# Inputs from the problem description
today = date(2019,12,30)

#
N, M = 6, 9 #Number of bonds and number of time periods
K = 100
T = np.array([0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2]) #Time periods in years
L_3M, L_6M = 0.01570161, 0.0198204
pi = np.array([102.33689177, 104.80430234, 105.1615306,
               105.6581905, 104.02899992, 101.82604116]) #Prices of the fixed rate bullet bonds

# Problem 2a
C = np.zeros((N, M))
#Cashflows show annual rate times principal times how many times coupon is paid in a year.
#1: Maturing December 30. 2020 (in one year) paying quarterly (0.25), 4% annually (0.04): 
C[0, :] = [0.04*K*0.25, 0.04*K*0.25, 0.04*K*0.25, 0.04*K*0.25, 0.04*K*0.25+K, 0, 0, 0, 0]
C[1, :] = [0.05*K*0.5, 0, 0.05*K*0.5, 0, 0.05*K*0.5+K, 0, 0, 0, 0]
C[2, :] = [0.05*K*0.5, 0, 0.05*K*0.5, 0, 0.05*K*0.5, 0, 0.05*K*0.5+K, 0, 0]
C[3, :] = [0.06*K*0.25, 0.06*K*0.25, 0.06*K*0.25, 0.06*K*0.25, 0.06*K*0.25, 0.06*K*0.25, 0.06*K*0.25+K, 0, 0]
C[4, :] = [0.05*K*0.25, 0.05*K*0.25, 0.05*K*0.25, 0.05*K*0.25, 0.05*K*0.25, 0.05*K*0.25, 0.05*K*0.25, 0.05*K*0.25, 0.05*K*0.25+K]
C[5, :] = [0.03*K*1, 0, 0, 0, 0.03*K*1, 0, 0, 0, 0.03*K*1+K]
print(f"2a - The cashflow matrix becomes:")
print(C)

```

## b) Find zero-coupon prices 
(Find the vector of zero coupon prices for all the times that you can based on the above information
and find the term structure of continuously compounded zero coupon spot rates (the yield curve).
Report the results and plot both curves in an appropriate diagram.)

Solve for **zero-coupon prices** $(p(0,T))$ on all reachable payment dates and compute the **continuously compounded spot rates** $R(T) = -\ln [p(0,T)/T]$.


```python
# Problem 2b
p, r = np.ones([M]), np.zeros([M]) #p stores the zero-coupon prices, r stores the spot rates. At first we assume all prices are 1 and all rates are 0.
p[1], p[2] = 1/(1+L_3M*T[1]), 1/(1+L_6M*T[2]) #Prices for first two maturities from 3M and 6M Libor rates. T[1]=0.25=3 months, T[2]=0.5=6 months
y = pi - p[0]*C[:,0] - p[1]*C[:,1] - p[2]*C[:,2] #Adjusted prices after removing known cashflows (we know the two LIBOR rates)
C_b = C[:,3:] #Remaining cashflows after removing known cashflows
p[3:M] = np.linalg.solve(C_b, y) #Solving the linear system of equations to find the remaining zero-coupon prices
for i in range(1, M): 
    r[i] = -np.log(p[i])/T[i] #Calculating the spot rates from the zero-coupon prices

print(f"2b - ZCB prices: {p}")
print(f"2b - spot rates: {r}")
```

## C) 3M forward rates 

Find 3M forward rates and plot these in the diagram from b).

```python
f_3M = np.zeros(M) 
f_3M[0]= r[0] #The first forward rate is equal to the first spot rate
for i in range(1,M): 
    f_3M[i] = -(np.log(p[i])-np.log(p[i-1]))/(T[i]-T[i-1]) #Calculating the 3M forward rates from the zero-coupon prices
print(f"2b - 3M Forward rates: {f_3M}")
```

## d) 2 year Floating Rate Bullet 

Find the price of a 2 year floating rate bullet note with principal 100 paying 6M LIBOR issued today. 

```python
pi_float = K*p[0] # A par floater (reset today, paying 6M LIBOR) is always worth its face value.
print(f"2d - The floating rate note trades at par: {pi_float}")
```

Result: The price of a newly issued floating rate bond will always be equal to the principal.

## e) Par Swap for 2-year interest rate swap 

Find the par swap for a 2-year interest rate swap paying semi-annual fixed coupons at annual rate R to the receiver and quarterly 3M LIBOR to the "Payer" issued today.

```python
R_swap = (1-p[8]/(0.5*(p[2]+p[4]+p[6]+p[8]))) #p[8] is the price of the 2 year ZCB, p[2], p[4], p[6] are the prices of the 6M, 1Y and 1.5Y ZCBs
#Each semi-annual coupon is weighted by 0.5 because they are paid every half year.
print(f"2e - par swap rate: {R_swap}")                    
```
